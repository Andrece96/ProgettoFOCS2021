### 4. For each ranking, consider only the most recent data point. For each university, compute the maximum difference between the rankings (e.g. for Aarhus University the value is 122-73=49). Notice that some rankings are expressed as a range


Questa query richiede di lavorare su tutti i ranking a mia disposizione, procedo a memorizzarli (ricaricando anche il *times_dataset*)

```python
times_dataset = pd.read_csv("https://raw.githubusercontent.com/Andrece96/ProgettoFOCS2021/main/timesData.csv")
times_dataset = times_dataset.iloc[times_dataset.groupby('university_name')['year'].idxmax()]
times_dataset = times_dataset[['world_rank', 'university_name']]
times_dataset.head(10)
```

```python
shangai_dataset = pd.read_csv("https://raw.githubusercontent.com/Andrece96/ProgettoFOCS2021/main/shanghaiData.csv")
shangai_dataset = shangai_dataset.iloc[shangai_dataset.groupby('university_name')['year'].idxmax()]
shangai_dataset = shangai_dataset[['world_rank', 'university_name']]
shangai_dataset.head(10)
```

```python
cwur_dataset = pd.read_csv("https://raw.githubusercontent.com/Andrece96/ProgettoFOCS2021/main/cwurData.csv")
cwur_dataset = cwur_dataset.iloc[cwur_dataset.groupby('institution')['year'].idxmax()]
cwur_dataset = cwur_dataset[['world_rank', 'institution']]
cwur_dataset.head(10)
```

Nel *cwur_dataset* la colonna *institution* rappresenta l'omologo di *university_name* degli altri due. Decido di rinominarla per uniformare i dati ed eseguire un JOIN su quell'attributo.

```python
cwur_dataset.rename(columns={"institution": "university_name"}, inplace = True)
cwur_dataset.head(3)
```

Unisco i dataset in funzione di *university_name*. Rinomino inoltre le colonne *world_rank* per riconoscere facilmente la provenienza dei dati

```python
times_dataset.rename(columns={"world_rank": "times_rank"}, inplace = True)
shangai_dataset.rename(columns={"world_rank": "shangai_rank"}, inplace = True)
cwur_dataset.rename(columns={"world_rank": "cwur_rank"}, inplace = True)
```

```python
global_ranking = pd.merge(times_dataset,shangai_dataset,on='university_name')
global_ranking = pd.merge(global_ranking,cwur_dataset,on='university_name')
global_ranking
```

1. alcuni ranking sono espressi come range numerico. Per risolvere questa eterogeneità tra i dati, sostituirò i range con il valor medio degli estremi. 
1. alcuni ranking contengono caratteri non numerici, come nel caso del *times_dataset* che vede alcuni valori preceduti dal carattere "=". Per omogeneizzare i dati rimuovo il carattere "=".

```python
global_ranking['times_rank'] = global_ranking['times_rank'].str.replace('=','')

global_ranking['times_MAX'] = np.where(global_ranking['times_rank'].str.contains('-'), 
                                         global_ranking['times_rank'].str.split(pat="-").str[-1], 
                                         global_ranking['times_rank'])

global_ranking['times_MIN'] = np.where(global_ranking['times_rank'].str.contains('-'), 
                                         global_ranking['times_rank'].str.split(pat="-").str[-2], 
                                         global_ranking['times_rank'])

global_ranking = global_ranking.astype({'times_MAX': 'int64', 'times_MIN': 'int64'})
global_ranking
```

Calcolo la media tra *times_MAX* e *times_MIN*, memorizzo il risultato nell'attributo *times_rank* e rimuovo le colonne ausiliarie *times_MAX* e *times_MIN* create in precedenza.

```python
global_ranking['times_rank'] = np.where(global_ranking['times_MAX']==global_ranking['times_MIN'], 
                                         global_ranking['times_rank'], 
                                         np.floor((global_ranking['times_MAX']+global_ranking['times_MIN'])/2))
del global_ranking['times_MAX'], global_ranking['times_MIN']
global_ranking
```

Ripeto la stessa procedura per l'attributo *shangai_rank*

```python
global_ranking['shangai_MAX'] = np.where(global_ranking['shangai_rank'].str.contains('-'), 
                                         global_ranking['shangai_rank'].str.split(pat="-").str[-1], 
                                         global_ranking['shangai_rank'])

global_ranking['shangai_MIN'] = np.where(global_ranking['shangai_rank'].str.contains('-'), 
                                         global_ranking['shangai_rank'].str.split(pat="-").str[-2], 
                                         global_ranking['shangai_rank'])

global_ranking = global_ranking.astype({'shangai_MAX': 'int64', 'shangai_MIN': 'int64'})
global_ranking
```

```python
global_ranking['shangai_rank'] = np.where(global_ranking['shangai_MAX']==global_ranking['shangai_MIN'], 
                                         global_ranking['shangai_rank'], 
                                         np.floor((global_ranking['shangai_MAX']+global_ranking['shangai_MIN'])/2))
del global_ranking['shangai_MAX'], global_ranking['shangai_MIN']
global_ranking
```

Gli attributi *times_rank* e *shangai_rank* sono ancora di tipo oggetto. Dovendo eseguire operazioni artimetiche tra questi, ne cambio il tipo.

```python
global_ranking.dtypes
```

```python
global_ranking = global_ranking.astype({'times_rank': 'int64', 'shangai_rank': 'int64'})
```

Creo le colonne ausiliarie *max_rank* e *min_rank* in cui riporto il massimo ed il minimo per tupla tra *times_rank*, *shangai_rank* e *cwur_rank*. Calcolo la differenza tra *max_rank* e *min_rank*, memorizzandola in un nuovo attributo, poi rimuovo le colonne ausiliarie.

```python
global_ranking['max_rank'] = global_ranking[['times_rank', 'shangai_rank', 'cwur_rank']].max(axis=1)
global_ranking['min_rank'] = global_ranking[['times_rank', 'shangai_rank', 'cwur_rank']].min(axis=1)
global_ranking['difference'] = global_ranking['max_rank'] - global_ranking['min_rank']
del global_ranking['max_rank'], global_ranking['min_rank']
global_ranking
```

### 5. Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country.


Procedo ad importare nuovamente il *times_dataset* selezionando solo i records più recenti

```python
times_dataset = pd.read_csv("https://raw.githubusercontent.com/Andrece96/ProgettoFOCS2021/main/timesData.csv")
times_dataset = times_dataset.iloc[times_dataset.groupby('university_name')['year'].idxmax()]
times_dataset.head(3)
```

Restringo il dataset agli attributi di mio interesse

```python
times_dataset = times_dataset[['country', 'num_students', 'female_male_ratio']]
times_dataset
```

Rimuovo le righe contenenti valori NaN ed il carattere "-", osservati nella colonna *female_male_ratio*

```python
times_dataset = times_dataset[(times_dataset['female_male_ratio'] != '-') & (times_dataset['female_male_ratio'].notnull())].copy()
times_dataset.head(15)
```

Gli attributi *num_students* e *female_male_ratio* sono di tipo oggetto. Occorre cambiarlo per eseguire operazioni aritmetiche tra questi.

```python
times_dataset.dtypes
```

Per quanto riguarda la colonna *female_male_ratio*, data la lunghezza dei valori presenti, immagino che siano tutti espressi come proporzione, utilizzando i caratteri " : " tra i due termini. Separo i termini della proporzione memorizzando ciascuno in una colonna ausiliaria.

```python
times_dataset.female_male_ratio.str.len().value_counts()
```

```python
times_dataset['female_rate'] = np.where(times_dataset['female_male_ratio'].str.contains(' : '), 
                                         times_dataset['female_male_ratio'].str.split(pat=" : ").str[-2], 
                                         times_dataset['female_male_ratio'])

times_dataset['male_rate'] = np.where(times_dataset['female_male_ratio'].str.contains(' : '), 
                                         times_dataset['female_male_ratio'].str.split(pat=" : ").str[-1], 
                                         times_dataset['female_male_ratio'])

times_dataset = times_dataset.astype({'female_rate': 'int64', 'male_rate': 'int64'})
times_dataset
```

Per quanto riguarda invece la colonna *num_students*, rimuovo il carattere ',' per ottenere valori solo numerici. Cambio anche qui il tipo di dato in numerico

```python
times_dataset['num_students'] = times_dataset['num_students'].str.replace(',','')
times_dataset = times_dataset.astype({'num_students': 'int64'})
times_dataset
```

```python
times_dataset.dtypes
```

Adesso posso determinare l'esatto numero di studenti e studentesse su ogni tupla

```python
times_dataset['female_count'] = round(times_dataset['num_students'] * times_dataset['female_rate'] * 0.01)
times_dataset['male_count'] = round(times_dataset['num_students'] * times_dataset['male_rate'] * 0.01)
times_dataset
```

Rimuovo le colonne inutili per rendere più chiaro il risultato.

```python
del times_dataset['num_students'], times_dataset['female_male_ratio'], times_dataset['female_rate'], times_dataset['male_rate']
times_dataset
```

Raggruppo i dati per nazione e calcolo il valore totale degli attributi *female_count* e *male_count*

```python
times_dataset_students_in_country = times_dataset.groupby('country')[['female_count', 'male_count']].sum()
times_dataset_students_in_country
```

### 6. Find the universities where the ratio between female and male is below the average ratio (computed over all universities)


Riprendo parte del codice eseguito per risolvere la richiesta precedente.

```python
times_dataset = pd.read_csv("https://raw.githubusercontent.com/Andrece96/ProgettoFOCS2021/main/timesData.csv")
times_dataset = times_dataset.iloc[times_dataset.groupby('university_name')['year'].idxmax()]
times_dataset = times_dataset[['university_name', 'num_students', 'country', 'female_male_ratio']]

times_dataset = times_dataset[(times_dataset['female_male_ratio'] != '-') & (times_dataset['female_male_ratio'].notnull())].copy()

times_dataset['female_rate'] = np.where(times_dataset['female_male_ratio'].str.contains(' : '), 
                                         times_dataset['female_male_ratio'].str.split(pat=" : ").str[-2],
                                        times_dataset['female_male_ratio'])

times_dataset['male_rate'] = np.where(times_dataset['female_male_ratio'].str.contains(' : '), 
                                         times_dataset['female_male_ratio'].str.split(pat=" : ").str[-1],
                                      times_dataset['female_male_ratio'])

times_dataset = times_dataset.astype({'female_rate': 'int64', 'male_rate': 'int64'})
del times_dataset['female_male_ratio']
times_dataset
```

Noto che una delle università presenta un *male_rate* pari a 0. Questo dato può compromettere il risultato di operazioni aritmetiche eseguite sui dati, quindi decido di escluderlo.

```python
times_dataset.loc[times_dataset['male_rate'] == 0]
```

```python
times_dataset = times_dataset[(times_dataset['male_rate'] != 0)].copy()
times_dataset
```

Per ogni università, calcolo lo specifico rapporto $\frac{femmine}{maschi}$ e lo memorizzo in un nuovo attributo, *computed_rate*

```python
times_dataset['computed_rate'] = times_dataset['female_rate'] / times_dataset['male_rate']
times_dataset
```

Calcolo il valor medio dell'attributo *computed_rate* e memorizzo anche questo dato in una colonna ausiliaria, *global_mean*

```python
times_dataset['global_mean'] = times_dataset['computed_rate'].mean()
times_dataset
```

Seleziono i records che presentano $\frac{femmine}{maschi} < global$_$mean$

```python
times_dataset = times_dataset[times_dataset['computed_rate'] < times_dataset['global_mean']].copy()
times_dataset
```

```python
cwur_dataset = pd.read_csv("https://raw.githubusercontent.com/Andrece96/ProgettoFOCS2021/main/cwurData.csv")
cwur_dataset = cwur_dataset[['world_rank', 'institution', 'year']]
cwur_dataset.rename(columns={"institution": "university_name"}, inplace = True)
cwur_dataset
```

Prima di unire i tre dataset limitatamente alle università presenti in ciascuno, rinomino l'attributo *world_rank*  in ciascuno per identificare facilmente la provenienza dei dati

```python
times_dataset.rename(columns={"world_rank": "times_rank"}, inplace = True)
shangai_dataset.rename(columns={"world_rank": "shangai_rank"}, inplace = True)
cwur_dataset.rename(columns={"world_rank": "cwur_rank"}, inplace = True)
global_ranking = pd.merge(times_dataset,shangai_dataset,on = ['university_name', 'year'])
global_ranking = pd.merge(global_ranking,cwur_dataset,on = ['university_name', 'year'])

global_ranking
```

Diversi valori degli attributi *times_rank* e *shangai_rank* sono espressi come range. Richiamo il codice eseguito per la richiesta 4.

```python
global_ranking['times_max'] = np.where(global_ranking['times_rank'].str.contains('-'), 
                                         global_ranking['times_rank'].str.split(pat="-").str[-1], 
                                         global_ranking['times_rank'])

global_ranking['times_min'] = np.where(global_ranking['times_rank'].str.contains('-'), 
                                         global_ranking['times_rank'].str.split(pat="-").str[-2], 
                                         global_ranking['times_rank'])

global_ranking = global_ranking.astype({'times_max': 'int64', 'times_min': 'int64'})

global_ranking['shangai_max'] = np.where(global_ranking['shangai_rank'].str.contains('-'), 
                                         global_ranking['shangai_rank'].str.split(pat="-").str[-1], 
                                         global_ranking['shangai_rank'])

global_ranking['shangai_min'] = np.where(global_ranking['shangai_rank'].str.contains('-'), 
                                         global_ranking['shangai_rank'].str.split(pat="-").str[-2], 
                                         global_ranking['shangai_rank'])

global_ranking = global_ranking.astype({'shangai_max': 'int64', 'shangai_min': 'int64'})

del global_ranking['times_rank'], global_ranking['shangai_rank']
global_ranking
```

Filtro i records prendendo solo quelli in cui ogni università è posizionata allo stesso modo tra i vari rankings

```python
global_ranking = global_ranking.loc[(global_ranking['cwur_rank'] >= global_ranking['times_min']) 
                                  & (global_ranking['cwur_rank'] >= global_ranking['shangai_min'])
                                  & (global_ranking['cwur_rank'] <= global_ranking['times_max']) 
                                  & (global_ranking['cwur_rank'] <= global_ranking['shangai_max'])]

global_ranking
```
